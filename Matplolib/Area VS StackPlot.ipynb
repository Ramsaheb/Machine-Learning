```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
area1 = [3, 2, 5, 4]
area2 = [1, 3, 2, 5]
area3 = [0, 3, 7, 5]

l = ['area1', 'area2', 'area3']
c = ['r', 'b', 'm']

plt.stackplot(x, area1, area2, area3, labels=l, colors=c, baseline='wiggle')

plt.legend(loc=2)
plt.show()
```

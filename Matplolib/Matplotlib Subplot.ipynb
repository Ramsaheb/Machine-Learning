```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]


plt.subplot(2, 2, 1)
plt.plot(x, y)

plt.subplot(2, 2, 2)
plt.pie([1])

plt.subplot(2, 2, 3)
plt.pie(x=[1, 2, 3, 4])


x1 = ['python', 'java', 'c++']
y1 = [60, 20, 20]
plt.subplot(2, 2, 4)
plt.bar(x1, y1)



plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
x = [30,20,50]
y = ['c++', 'java', 'python']
ex = [0.3, 0.4, 0.3,]
c = ['r', 'y', 'blue']

plt.pie(x, labels=y, #explode=ex, 
        colors=c, autopct='%0.1f%%', shadow=True, radius=1,
        labeldistance=1.1, startangle=80, textprops={'fontsize':15},
       counterclock=False, #wedgeprops={'linewidth':5, 'edgecolor':'g'},
        rotatelabels=False, 
       )

plt.title('Popularity')
plt.legend()
plt.show()
```

```python
x = [30, 20, 50]
x1 = [10, 30, 60]
y = ['c++', 'java', 'python']
c = ['r', 'y', 'blue']

plt.pie(x, labels=y, radius=1)
plt.pie(x1, radius=0.5, colors=c)

plt.show()
```

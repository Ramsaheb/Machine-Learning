```python
import matplotlib.pyplot as plt
```

```python
day = [1, 2, 3, 4, 5, 6]
no = [3, 4, 2, 2, 5, 2]
no1 = [2, 3, 4, 5, 6, 7]
colors1 = [10, 26, 73, 54, 53, 73]
colors2 = ['r', 'y', 'b', 'b', 'g', 'g'][::-1]
sizes = [100, 100, 100, 100, 100, 100,]
plt.scatter(day, no, c=colors1, s=sizes, alpha=0.9, marker='*',
            edgecolor=colors2, linewidth=4, cmap='viridis')
plt.scatter(day, no1, c=colors1, s=sizes, alpha=0.9, marker='*',
            edgecolor=colors2, linewidth=4, cmap='viridis')
t = plt.colorbar()
t.set_label('colorbar', size=20)
plt.title('Scatter Plot', size=30)
plt.xlabel('Day', size=15)
plt.ylabel('No.', size=15)
plt.show()
```

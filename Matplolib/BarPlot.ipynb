**BAR PLOT**

```python
import matplotlib.pyplot as plt
import numpy as np

x = ['python', 'java', 'c++']
y = [100, 72, 64]
z = [50, 30, 15]
width = 0.2
p = np.arange(len(x))
p1 = [j + width for j in p]

plt.title('Demo', fontsize=20)
plt.xlabel('Languages', fontsize=10)
plt.ylabel('Popularity', fontsize=10)

c = ['r', 'b', 'c']
e = ['r', 'b', 'c'][::-1]

plt.bar(p, y, width, color='r' , #align='center', #edgecolor='r',#
        #linewidth=1#, #linestyle=':',#
        #alpha=0.9,#
        label='wscube'
       )
plt.bar(p1, z, width, color='y', #align='center', #edgecolor='y',#
        #linewidth=1#, #linestyle=':',#
        #alpha=0.9,#
        label='wscube'
       )

plt.xticks(p+width/2, x, rotation=10)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
var = np.linspace(1, 10, 20).reshape(4, 5)
var
```

```python
sns.heatmap(data=var)
plt.show()
```

```python
data = sns.load_dataset('anscombe')
df = data.drop(columns='dataset', axis=1)
df
```

```python
a = np.array(['a', 'b',
              'c', 'd',
              'e', 'f'])
sns.heatmap(data=df.head(3), cmap='PuOr', annot=a, fmt='s')
plt.show()
```
